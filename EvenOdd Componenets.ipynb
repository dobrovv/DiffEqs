```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def EvenComponent(x, f):
    return (f(x)+f(-x))/2;

def OddComponent(x, f):
    return (f(x)+(-f(-x)))/2;

f = lambda x: np.sin(x**2) * np.abs(x)

x = np.arange(-3, 3, 0.1)

plt.figure(1)
plt.plot(x, f(x))

plt.figure(2)
plt.subplot(211)
plt.plot(x,EvenComponent(x,f))
plt.title("Even Component")

plt.subplot(212)
plt.plot(x,OddComponent(x,f))
plt.title("Odd Component")
plt.tight_layout()


    
```
